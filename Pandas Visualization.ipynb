```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/andyokwon/Downloads/Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available) # Able to change visual in matplotlib
plt.style.use('seaborn-v0_8-paper')
```

```python
df.plot() # Line plot by default
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(subplots = True, kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar') #Pull out one column
```

```python
df.plot.barh(stacked = True) #Horizontal bar chart
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating') #Must specify x and y axis for scatter plots
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Red') # s = size, c = color
```

```python
df.plot.hist()
```

```python
df.plot.hist(bins = 50) # bins = distribution 
```

```python
df.plot.area()
```

```python
df.plot.area(figsize = (10,5)) 
```
